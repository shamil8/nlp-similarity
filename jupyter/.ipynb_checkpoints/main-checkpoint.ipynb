```python
import pandas as pd
```

```python
df = pd.read_csv('../data/tasks-from-mover.csv')
```

```python
user_id = '2be0ccf2-3608-11ea-9dc1-0242c0a85009'
words_nlp = [
    "гидравлический",
    "люди",
]
```

```python
df.head(10)
```

```python
# only once!
titles = df['name'].str.lower()
```

```python
titles
```

```python
tasks = []
tmp = []
```

```python
for idx, words in enumerate(titles):
    for word in str(words).split():
        if word in words_nlp:
            if idx in tmp:
                for index, [task_idx, rating] in enumerate(tasks):
                    if task_idx == idx:
                        tasks[index][1] += 1  # [index][1] - it's a rating!
            else:
                tmp.append(idx)
                tasks.append([idx, 1])
```

```python
tasks
```

```python
data = []
```

```python
for idx, rating in tasks:
    owner_id, times = df.loc[idx, ['owner_id', 'times']]
    # space for adding params
    if owner_id == user_id:
        rating += 2
    
    # END space for adding params
    
    data.append([rating, times])
```
